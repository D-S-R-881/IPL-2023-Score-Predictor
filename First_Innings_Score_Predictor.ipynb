```python
stadiums = ['Wankhede Stadium','Eden Gardens','Arun Jaitley Stadium','M.Chinnaswamy Stadium','MA Chidambaram Stadium','Narendra Modi Stadium',
           'Rajiv Gandhi International Stadium','Punjab Cricket Association IS Bindra Stadium','Sawai Mansingh Stadium',
           'Himachal Pradesh Cricket Association Stadium','Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium','Barsapara Cricket Stadium']
```

```python
df_new = df_new[df_new['venue'].isin(stadiums)]
df_new['venue'].value_counts()
```

```python
df_new['current_score'] = df_new.groupby('match_id').cumsum()['runs_scored']
```

```python
df_new
```

```python
df_new['over'] = df_new['ball'].apply(lambda x:str(x).split(".")[0])
df_new['ball_no'] = df_new['ball'].apply(lambda x:str(x).split(".")[1])
df_new
```

```python
df_new['balls_bowled'] = (df_new['over'].astype('int')*6) + df_new['ball_no'].astype('int')
df_new
```

```python
df_new['balls_left'] = 120-df_new['balls_bowled']
df_new['balls_left'] = df_new['balls_left'].apply(lambda x:0 if x<0 else x)
df_new
```

```python
df_new['player_dismissed'] = df_new['player_dismissed'].apply(lambda x:0 if x==0 else 1)
df_new
```

```python
df_new['player_dismissed'] = df_new['player_dismissed'].astype('int')
df_new['player_dismissed'] = df_new.groupby('match_id').cumsum()['player_dismissed']
df_new['wickets_left'] = 10 - df_new['player_dismissed']
```

```python
df_new
```

```python
df_new['current_run_rate'] = (df_new['current_score']*6)/df_new['balls_bowled']
df_new
```

```python
groups = df_new.groupby('match_id')
match_ids = df_new['match_id'].unique()
last_five = []

for id in match_ids:
  last_five.extend(groups.get_group(id).rolling(window = 30).sum()['runs_scored'].values.tolist())
```

```python
df_new['last_five'] = last_five
df_new
```

```python
final_df = df_new.groupby('match_id').sum()['runs_scored'].reset_index().merge(df_new,on='match_id')
```

```python
final_df
```

```python
final_df=final_df[['batting_team','bowling_team','venue','current_score','balls_left','wickets_left','current_run_rate','last_five','runs_scored_x']]
final_df
```

```python
final_df.dropna(inplace=True)
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.head()
```

```python
X = final_df.drop(columns=['runs_scored_x'])
y = final_df['runs_scored_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','venue'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=10))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
